### 3.1.1 | CorridorID

* CorridorID / corridorName berhubungan satu sama lain. Pengisian missing values dilakukan dengan metode pencocokan antara satu dengan yang lain.

```python
corridor_id_to_name = TJ.dropna(subset=['corridorID', 'corridorName']).set_index('corridorID')['corridorName'].to_dict()
corridor_name_to_id = TJ.dropna(subset=['corridorID', 'corridorName']).set_index('corridorName')['corridorID'].to_dict()

TJ['corridorID'] = TJ.apply(
    lambda row: corridor_name_to_id[row['corridorName']] if pd.isna(row['corridorID']) and row['corridorName'] in corridor_name_to_id else row['corridorID'],
    axis=1
)

TJ['corridorName'] = TJ.apply(
    lambda row: corridor_id_to_name[row['corridorID']] if pd.isna(row['corridorName']) and row['corridorID'] in corridor_id_to_name else row['corridorName'],
    axis=1
)
```

```python
msno.matrix(TJ[missing_coloumns])
```

* Jumlah missing values pada baris yang kosong dari corridorID dan corridorName sudah sama.

### 3.1.2 | tapOutStopsName & tapOutStops

* tapOutStopsName / tapOutStops berhubungan satu sama lain. Pengisian missing values dilakukan dengan metode pencocokan antara satu dengan yang lain.

```python
tap_out_stops_to_name = TJ.dropna(subset=['tapOutStops', 'tapOutStopsName']).set_index('tapOutStops')['tapOutStopsName'].to_dict()
tap_out_name_to_stops = TJ.dropna(subset=['tapOutStops', 'tapOutStopsName']).set_index('tapOutStopsName')['tapOutStops'].to_dict()

TJ['tapOutStops'] = TJ.apply(
    lambda row: tap_out_name_to_stops[row['tapOutStopsName']] if pd.isna(row['tapOutStops']) and row['tapOutStopsName'] in tap_out_name_to_stops else row['tapOutStops'],
    axis=1
)

TJ['tapOutStopsName'] = TJ.apply(
    lambda row: tap_out_stops_to_name[row['tapOutStops']] if pd.isna(row['tapOutStopsName']) and row['tapOutStops'] in tap_out_stops_to_name else row['tapOutStopsName'],
    axis=1
)
```

### 3.1.3 | tapInStops

* tapInStopsName / tapInStops berhubungan satu sama lain. Pengisian missing values dilakukan dengan metode pencocokan antara satu dengan yang lain.

```python
tap_in_stops_to_name = TJ.dropna(subset=['tapInStops', 'tapInStopsName']).set_index('tapInStops')['tapInStopsName'].to_dict()
tap_in_name_to_stops = TJ.dropna(subset=['tapInStops', 'tapInStopsName']).set_index('tapInStopsName')['tapInStops'].to_dict()

TJ['tapInStops'] = TJ.apply(
    lambda row: tap_in_name_to_stops[row['tapInStopsName']] if pd.isna(row['tapInStops']) and row['tapInStopsName'] in tap_in_name_to_stops else row['tapInStops'],
    axis=1
)

TJ['tapInStopsName'] = TJ.apply(
    lambda row: tap_in_stops_to_name[row['tapInStops']] if pd.isna(row['tapInStopsName']) and row['tapInStops'] in tap_in_stops_to_name else row['tapInStopsName'],
    axis=1
)
```

### 3.1.4 | PayAmount

```python
Mikrotrans = TJ[TJ['payAmount'] == 0]['corridorID'].unique()
Mikrotrans
```

```python
Transjakarta = TJ[TJ['payAmount'] == 3500]['corridorID'].unique()
Transjakarta
```

```python
Royaltrans = TJ[TJ['payAmount'] == 20000]['corridorID'].unique()
Royaltrans
```

Dari sini, dapat disimpulkan bahwa corridorID dengan payAmount 0, 3500, dan 20000 mempunyai unique values masing-masing. Maka dari itu kita bisa menggunakan corridorID untuk menentukan/ mengisi missing values pada payAmount.

```python
def fill_pay_amount(row):
    if row['corridorID'] in Royaltrans and pd.isna(row['payAmount']):
        return 20000
    elif row['corridorID'] in Mikrotrans and pd.isna(row['payAmount']):
        return 0
    elif row['corridorID'] in Transjakarta and pd.isna(row['payAmount']):
        return 3500
    elif pd.isna(row['payAmount']):
        return 3500   
    else:
        return row['payAmount']

TJ['payAmount'] = TJ.apply(fill_pay_amount, axis=1)
```

### 3.3.2 | Menambahkan Kolom Generation berdasarkan Umur

Menambahkan kolom kategori umur, yang dibagi menjadi 5 berdasarkan tahun lahir, yaitu :

1. Silent Generation    ( tahun lahir < 1946 )
2. Baby Boomer          ( 1946 <= tahun lahir <= 1964 )
3. Generation X         ( 1965 <= tahun lahir <= 1980 )
4. Millenial            ( 1981 <= tahun lahir <= 1996 )
5. Generation Z         ( 1997 <= tahun lahir <= 2011 )
6. Generation A         ( 2012 <= tahun lahir <= Sekarang )

```python
TJ_clean['payCardBirthDate'].sort_values()
```

```python
def determine_generation(birth_year):
    if birth_year < 1946:
        return 'Silent Generation'
    elif 1946 <= birth_year <= 1964:
        return 'Baby Boomer'
    elif 1965 <= birth_year <= 1980:
        return 'Generation X'
    elif 1981 <= birth_year <= 1996:
        return 'Millennial'
    elif 1997 <= birth_year <= 2011:
        return 'Generation Z'
    else:
        return 'Generation Alpha'

TJ_clean['Generation'] = TJ_clean['payCardBirthDate'].apply(determine_generation)
TJ_clean
```

```python
import pandas as pd
import folium

def create_map_for_corridor(TJ_clean, corridor_id):
    map_jakarta = folium.Map(location=[-6.175392, 106.827752], zoom_start=12)

    color_map = {
        '1T': 'blue',
        'S21': 'green',
        'JIS3': 'red',
        'JAK.06': 'purple',
        '11P': 'orange',
        '8C': 'darkred',
        '2E': 'lightblue',
        '9D': 'darkgreen',
        'M7B': 'lightgreen',
        'JAK.72': 'pink'
    }
    tap_in_counts = TJ_clean[TJ_clean['corridorID'] == corridor_id].groupby(['tapInStopsName', 'tapInStopsLat', 'tapInStopsLon']).size().reset_index(name='tapInCount')
    tap_out_counts = TJ_clean[TJ_clean['corridorID'] == corridor_id].groupby(['tapOutStopsName']).size().reset_index(name='tapOutCount')

    combined_counts = pd.merge(tap_in_counts, tap_out_counts, left_on='tapInStopsName', right_on='tapOutStopsName', how='outer')

    combined_counts = combined_counts.dropna(subset=['tapInStopsLat', 'tapInStopsLon'])

    total_stops = combined_counts.shape[0]
    print(f"Jumlah total halte untuk corridor {corridor_id}: {total_stops}")

    low_count_stops = 0

    for _, row in combined_counts.iterrows():
        tap_in_count = row['tapInCount'] if 'tapInCount' in row else 0
        tap_out_count = row['tapOutCount'] if 'tapOutCount' in row else 0

        if pd.isna(tap_out_count):
            tap_out_count = 0

        total_count = tap_in_count + tap_out_count

        marker_color = 'black' if total_count < 2 else color_map.get(corridor_id, 'blue')

        folium.Marker(
            location=[row['tapInStopsLat'], row['tapInStopsLon']],
            popup=f"{row['tapInStopsName']} ({corridor_id}): Tap In = {tap_in_count}, Tap Out = {tap_out_count}",
            icon=folium.Icon(color=marker_color)
        ).add_to(map_jakarta)

        if total_count < 2:
            low_count_stops += 1
            print(f"Halte dengan total count tapIn dan tapOut < 2: {row['tapInStopsName']} (Tap In: {tap_in_count}, Tap Out: {tap_out_count})")

    print(f"Jumlah total halte dengan tapIn dan tapOut di bawah 2 untuk corridor {corridor_id}: {low_count_stops}")

    return map_jakarta

map_jakarta
```

```python
import pandas as pd

# Data corridor
data = {
    "corridorID": ["JAK.06", "11P", "8C", "9D", "JAK.72"],
    "jumlah_halte_awal": [25, 25, 24, 30, 26],
    "jumlah_halte_tidak_efisien": [4, 7, 4, 6, 5],
    "waktu_tempuh_awal": [60, 78, 98, 74, 87],
    "waktu_pemberhentian": [2, 2, 1.5, 1.5, 2],
}

df = pd.DataFrame(data)

df["jumlah_halte_akhir"] = df["jumlah_halte_awal"] - df["jumlah_halte_tidak_efisien"]
df["waktu_tempuh_akhir"] = df["waktu_tempuh_awal"] - (df["jumlah_halte_tidak_efisien"] * df["waktu_pemberhentian"])

df["waktu_efisiensi"] = (df["jumlah_halte_tidak_efisien"] * df["waktu_pemberhentian"])

df["persentase_efisiensi"] = (df["waktu_efisiensi"] / df["waktu_tempuh_awal"]) * 100

df[["corridorID", "jumlah_halte_awal", "jumlah_halte_tidak_efisien", "jumlah_halte_akhir", "waktu_tempuh_awal", "waktu_tempuh_akhir", "persentase_efisiensi"]]
```
